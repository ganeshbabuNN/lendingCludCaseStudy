```python
# Grade

crosstab = pd.crosstab(loan_df['grade'], loan_df['loan_status'],margins=True)
crosstab['Probability_Charged Off'] = round((crosstab['Charged Off']/crosstab['All']),3)
crosstab = crosstab[0:-1]
linePlot = crosstab[['Probability_Charged Off']]
ax = linePlot.plot(figsize=(10,8), marker='o',color = 'b')

aggre= loan_df.groupby(['grade','loan_status'])['grade'].agg(Frequency='count')
aggre.Frequency.unstack().plot.bar(stacked=True, ax = ax, secondary_y=True)
plt.title("Grade and its probability of being Charged Off")

## Subgrade
crosstab = pd.crosstab(loan_df['sub_grade'], loan_df['loan_status'],margins=True)
crosstab['Probability_Charged Off'] = round((crosstab['Charged Off']/crosstab['All']),3)
crosstab = crosstab[0:-1]
linePlot = crosstab[['Probability_Charged Off']]
ax = linePlot.plot(figsize=(10,8), marker='o',color = 'b')

aggre= loan_df.groupby(['sub_grade','loan_status'])['sub_grade'].agg(Frequency='count')
aggre.Frequency.unstack().plot.bar(stacked=True, ax = ax, secondary_y=True)
plt.title("Sub Grade and its probability of being Charged Off")
```

<b>Inference - Grade/SubGrade (Bivariate Analysis):</b>
- As the Grade increases, the probability of charge off is also increased.

```python
## Subgrade
crosstab = pd.crosstab(loan_df['home_ownership'], loan_df['loan_status'],margins=True)
crosstab['Probability_Charged Off'] = round((crosstab['Charged Off']/crosstab['All']),3)
crosstab = crosstab[0:-1]
linePlot = crosstab[['Probability_Charged Off']]
ax = linePlot.plot(figsize=(10,8), marker='o',color = 'b')

aggre= loan_df.groupby(['home_ownership','loan_status'])['home_ownership'].agg(Frequency='count')
aggre.Frequency.unstack().plot.bar(stacked=True, ax = ax, secondary_y=True)
plt.title("Home ownership and its probability of being Charged Off")
plt.figure(figsize=(20, 11))
plt.show()
```

<b> Inference - Home Ownership (Bivariate Analysis):</b>
